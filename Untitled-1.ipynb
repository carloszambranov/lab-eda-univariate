```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy import stats

# Load data
file_name = 'amz_uk_price_prediction_dataset.csv'
if os.path.exists(file_name):
    df = pd.read_csv(file_name)
else:
    df = pd.read_csv(os.path.expanduser('~/Desktop/') + file_name)

# Find column names dynamically
cat_col = [c for c in df.columns if 'category' in c.lower()][0]
price_col = [c for c in df.columns if 'price' in c.lower()][0]
rating_col = [c for c in df.columns if 'rating' in c.lower() or 'stars' in c.lower()][0]

# Clean price column
prices = df[price_col].copy()
if prices.dtype == 'object':
    prices = prices.astype(str).str.replace('£', '').str.replace(',', '')
    prices = pd.to_numeric(prices, errors='coerce')
prices = prices.dropna()

# Clean ratings
ratings = df[rating_col].dropna()
if ratings.max() > 5:
    ratings = ratings / 20 if ratings.max() <= 100 else ratings / 100

# PART 1: Frequency tables
category_freq = df[cat_col].value_counts()
top5_categories = category_freq.head(5)
top10_categories = category_freq.head(10)

# PART 1: Visualizations
plt.figure(figsize=(12, 6))
plt.bar(range(len(top10_categories)), top10_categories.values)
plt.xticks(range(len(top10_categories)), top10_categories.index, rotation=45)
plt.title('Top 10 Categories')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(top5_categories.values, labels=top5_categories.index, autopct='%1.1f%%')
plt.title('Top 5 Categories')
plt.show()

# PART 2: Price statistics
price_mean = prices.mean()
price_median = prices.median()
price_mode = prices.mode().iloc[0]
price_std = prices.std()
price_range = prices.max() - prices.min()

print(f"Price Mean: {price_mean:.2f}, Median: {price_median:.2f}, Mode: {price_mode:.2f}")
print(f"Price Std: {price_std:.2f}, Range: {price_range:.2f}")

# PART 3: Ratings statistics
rating_mean = ratings.mean()
rating_median = ratings.median()
rating_mode = ratings.mode().iloc[0]
rating_std = ratings.std()
rating_range = ratings.max() - ratings.min()
rating_skew = stats.skew(ratings.dropna())
rating_kurtosis = stats.kurtosis(ratings.dropna())

print(f"Rating Mean: {rating_mean:.2f}, Median: {rating_median:.2f}, Mode: {rating_mode:.2f}")
print(f"Rating Std: {rating_std:.3f}, Skewness: {rating_skew:.3f}, Kurtosis: {rating_kurtosis:.3f}")

# Rating histogram
plt.figure(figsize=(10, 6))
plt.hist(ratings, bins=20, edgecolor='black')
plt.title('Rating Distribution')
plt.show()
```
